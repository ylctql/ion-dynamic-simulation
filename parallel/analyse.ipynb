```python
import numpy as np
import matplotlib.pyplot as plt

N = 1000
start = 0
rounds = 40
dir_prefix = '../data_cache/60electrodes_tiny/'
dir_name = dir_prefix + 'iter%d/'%(N)

UL = np.load(dir_name + 'UL_adam_%d-%d.npy'%(start, rounds))
L = UL[:, -1]
plt.plot(range(start, rounds), L)
plt.xlabel('Iteration rounds')
plt.ylabel('Loss(um)')
plt.title('N=%d, adam, 10us/round'%N)
plt.show()
```

```python
# Find the minimum loss and corresponding U
min_index = np.argmin(L)
min_loss = L[min_index]
optimal_U = UL[min_index, :-1]
print("Optimal iteration: ", start + min_index)
print("Minimum loss: ", min_loss)
print("Optimal U: ", optimal_U)
```
