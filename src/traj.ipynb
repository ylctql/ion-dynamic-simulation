```python
# g=0.1
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation
from pathlib import Path

N = 150
t0 = 37.443
g = 0.1
dt = 9.022389064166401e-03
ts = t0
tf = 430

dir_name = f"../data_cache/traj/N={N:d},t0={t0:.3f},g={g:.10g}/r/"
N_files = len(list(Path(dir_name).glob("*.npy")))
time_stamps = t0 + np.arange(N_files) * dt
mask = (time_stamps<tf) & (time_stamps>ts)
N_steps = time_stamps[mask].shape[0]
traj = np.zeros((N_steps, N, 3))
for file_id, filename in enumerate(Path(dir_name).glob("*.npy")):
    if file_id >= N_steps:
        break
    filename = dir_name + filename.name
    traj[file_id] = np.load(filename)


# 
fig, ax = plt.subplots(figsize=(20, 8))
for i in range(N):
    ax.plot(traj[:, i, 2], traj[:, i, 0], 
             lw=1.5, alpha=0.7)
ax.scatter(traj[0, :, 2], traj[0, :, 0], s=5, c='b')
ax.scatter(traj[-1, :, 2], traj[-1, :, 0], s=5, c='r')
ax.set_aspect('equal')    

plt.title(f'{N:d} ions, g={g:.10g}, ts={ts:.6g}us, tf={tf:.6g}us')
plt.xlabel('z(um)')
plt.ylabel('x(um)')
plt.grid(True, alpha=0.3)
plt.axis('equal')
plt.legend()
plt.show()
```

```python
# 
r_eq = np.mean(traj, axis=0)
plt.scatter(r_eq[:, 2], r_eq[:, 0], s=5, alpha=0.7)
plt.gca().set_aspect('equal')
```
