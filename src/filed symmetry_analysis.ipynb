```python
#y方向对称性
%load_ext autoreload
%autoreload 2
import configure
import os
import numpy as np

dirname = os.getcwd()
flag_smoothing=True #是否对导入的电势场格点数据做平滑化，如果True，那么平滑化函数默认按照下文def smoothing(data)
filename=os.path.join(dirname, "../../data/bilayer_both_x50y250z200_dx2dy5dz5.csv")
basis_filename=os.path.join(dirname, "bilayer_basis.json")#文件名：自定义Basis设置 #可以理解为一种基矢变换，比如"U1"相当于电势场组合"esbe1"*0.5+"esbe1asy"*-0.5
config_name = "bilayer"
basis = configure.Data_Loader(filename, basis_filename, flag_smoothing)
basis.loadData()
def y_sym(key):
    plus = basis.load_basis(key+"+").reshape((len(basis.x), len(basis.y), len(basis.z)))*basis.dV
    minus = basis.load_basis(key+"-").reshape((len(basis.x), len(basis.y), len(basis.z)))*basis.dV
    print(f"{key} max", np.abs((plus-minus[:, ::-1, :])).max(), "min", np.abs((plus-minus[:, ::-1, :])).min(), "mean", np.abs((plus-minus[:, ::-1, :])).mean())#使U4-在y方向反号，则理论上应该完全一样
key_ls = ["RF", "U1", "U2", "U3", "U4", "U5", "U6", "U7"]
for key in key_ls:
    y_sym(key)
```

```python
#z方向对称性
%load_ext autoreload
%autoreload 2
import configure
import os
import numpy as np

dirname = os.getcwd()
flag_smoothing=True #是否对导入的电势场格点数据做平滑化，如果True，那么平滑化函数默认按照下文def smoothing(data)
filename=os.path.join(dirname, "../../data/bilayer_both_x50y250z200_dx2dy5dz5.csv")
basis_filename=os.path.join(dirname, "bilayer_basis.json")#文件名：自定义Basis设置 #可以理解为一种基矢变换，比如"U1"相当于电势场组合"esbe1"*0.5+"esbe1asy"*-0.5
config_name = "bilayer"
basis = configure.Data_Loader(filename, basis_filename, flag_smoothing)
basis.loadData()
def z_sym_item(key:str):
    V = basis.load_basis(key).reshape((len(basis.x), len(basis.y), len(basis.z)))*basis.dV
    print(f"{key} max", np.abs((V-V[:, :, ::-1])).max(), "min", np.abs((V-V[:, :, ::-1])).min(), "mean", np.abs((V-V[:, :, ::-1])).mean())
def z_sym(key_ls:list):
    l = len(key_ls)
    for i in range(l//2+1):
        Vm = basis.load_basis(key_ls[i]).reshape((len(basis.x), len(basis.y), len(basis.z)))*basis.dV
        Vp = basis.load_basis(key_ls[l-i-1]).reshape((len(basis.x), len(basis.y), len(basis.z)))*basis.dV
        print(f"{key_ls[i]}&{key_ls[l-i-1]} max", np.abs((Vm-Vp[:,:,::-1])).max(), "min", np.abs((Vm-Vp[:,:,::-1])).min(), "mean", np.abs((Vm-Vp[:,:,::-1])).mean())
key_ls = ["U1", "U2", "U3", "U4", "U5", "U6", "U7"]
key_ls_p = [key+"+" for key in key_ls]
key_ls_m = [key+"-" for key in key_ls]
z_sym_item("RF+")
z_sym_item("RF-")
z_sym(key_ls)
z_sym(key_ls_p)
z_sym(key_ls_m)
```
